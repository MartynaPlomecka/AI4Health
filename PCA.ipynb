```python
import scipy.io
import pandas as pd
import numpy as np
```

```python
data = scipy.io.loadmat('data/x.mat')
data['x'].shape
```

```python
beh = pd.read_csv("data/behaviorals.csv")
foof = pd.read_csv("data/foof2features.csv")
```

```python
df = pd.DataFrame(data['x'].reshape((data['x'].shape[0], -1)))
df['IDs'] = foof['id']
df
```

```python
df2 = pd.merge(df, beh[['DX_01', 'IDs']], on='IDs', how='inner')
df2
```
